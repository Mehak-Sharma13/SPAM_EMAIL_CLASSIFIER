```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('email.csv', encoding ='latin-1')
```

```python
print(data.head(7))
```

```python
print(data.shape)
```

```python
print(data.isnull().sum())
```

```python
data.drop_duplicates(inplace=True)
```

```python
print(data.shape)
```

```python
data['Category']=data['Category'].replace(['ham','spam'],['not spam','spam'])
```

```python
print(data.head(7))
```

```python
M=data['Message']
C=data['Category']
```

```python
(M_train,M_test,C_train,C_test)=train_test_split(M,C,test_size=0.2)
```

```python
cv=CountVectorizer(stop_words='english')
cv.fit_transform(M_train)
features=cv.fit_transform(M_train)
```

```python
model=MultinomialNB()
model.fit(features,C_train)
```

```python
features_test=cv.transform(M_test)
```

```python
print(model.score(features_test,C_test))
```

```python
mess=cv.transform(['congratulations,you won a lottery']).toarray()
result=model.predict(mess)
print(result)
```
